```python
import numpy as np
from scipy import stats
```

```python
# Given data
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
data
```

```python
# Sample statistics
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)   
```

```python
# Standard Error
SE = std / np.sqrt(n)

# t critical value for 99% CI (two-tailed)
t_crit = stats.t.ppf(1 - 0.005, df=n-1)

# Margin of Error
ME = t_crit * SE

# Confidence Interval
lower = mean - ME
upper = mean + ME
```

```python
print("Sample Mean:", mean)
print("Sample Std Dev:", std)
print("t Critical Value:", t_crit)
print("Margin of Error:", ME)
print("99% Confidence Interval: (", lower, ",", upper, ")")
```

```python
import numpy as np
from math import sqrt
from scipy.stats import norm

# Data
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
# Known population standard deviation
sigma = 0.2
n = len(data)

mean = np.mean(data)

# Standard Error
SE = sigma / np.sqrt(n)

# Z-critical for 99% CI
z_crit = norm.ppf(1 - 0.005)

# Margin of Error
ME = z_crit * SE

lower = mean - ME
upper = mean + ME
```

```python
print("99% CI (σ known):", (lower, upper))
```
