```python
# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print(" list:", numbers)
```

```python
# append()
numbers.append(60)
print("After append:", numbers)
```

```python
# extend()
numbers.extend([70, 80])
print("After extend:", numbers)
```

```python
# insert()
numbers.insert(2, 25)   # insert 25 at index 2
print("After insert:", numbers)
```

```python
# remove()
numbers.remove(40)      # remove value 40
print("After remove:", numbers)
```

```python
# pop()
popped_value = numbers.pop()  # removes last element
print("After pop():", numbers)
print("Popped value:", popped_value)
```

```python
# index()
index_of_25 = numbers.index(25)
print("Index of 25:", index_of_25)
```

```python
# count()
count_of_20 = numbers.count(20)
print("Count of 20:", count_of_20)
```

```python
# sort()
numbers.sort()
print("After sort():", numbers)
```

```python
# reverse()
numbers.reverse()
print("After reverse():", numbers)
```

```python
# clear()
numbers.clear()
print("After clear():", numbers)
```

```python
import pandas as pd
import numpy as np

# Creating a DataFrame 
data = {
    'Name': ['Aman', 'Sameer', 'Tanveer', 'Aman', 'Sara', 'Ali', 'Sameer', 'Meera', 'Suraj', 'Tanveer'],
    'Age': [25, 30, 28, 25, 22, 35, 30, 29, 40, 28],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Delhi', 'Pune', 'Chennai', 'Mumbai', 'Delhi', 'Kolkata', 'Delhi'],
    'Salary': [50000, 60000, np.nan, 50000, 45000, 70000, 60000, 58000, np.nan, 52000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# 2. Display info
df.info()
```

```python
df.describe()
```

```python
# 4. Select 'Name' and 'City' columns
print(df[['Name', 'City']])
```

```python
# 5. Drop 'City' column
df_dropped = df.drop(columns=['City'])
```

```python
df_dropped
```

```python
# 6. Fill missing values in Salary with mean
mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
```

```python
df
```

```python
# 7. Remove duplicate rows
df_unique = df.drop_duplicates()
```
