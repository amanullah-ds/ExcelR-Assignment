```python
#Prime Number
```

```python
num = int(input("Enter a number: "))

# Check if the number is prime

if num <= 1:
    print(num, "is not a prime number.")
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(num, "is not a prime number.")
            break
    else:
        print(num, "is a prime number.")
```

```python
#Product of Random Numbers
```

```python
import random

# Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user to calculate the product
print(f"What is the product of {num1} and {num2} ?")
user_answer = int(input("Enter your answer: "))

# Check if the user's answer is correct
correct_answer = num1 * num2

if user_answer == correct_answer:
    print("Your answer is correct.")
else:
    print("your answer is incorrect")
```

```python
 #Squares of Even/Odd Numbers
for num in range(100, 201):
    if num % 2 == 0:  # Check if the number is even
        square = num ** 2
        print(f"The square of {num} is {square}")
```

```python
#Hospital Billing System
```

```python
# Given data
room_charge_per_day = 2000
consultation_fee = 1500
lab_test_cost = 300
medicine_charges = 2400

# Patient details
days_stayed = 4
lab_tests = 3
```

```python
# Calculations
room_total = room_charge_per_day * days_stayed
lab_total = lab_test_cost * lab_tests

total_before_discount = room_total + consultation_fee + lab_total + medicine_charges
```

```python
# Apply discount if total exceeds 10000
if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
    discount = 0

final_bill = total_before_discount - discount
```

```python
# Output
print("Total before discount:", total_before_discount)
print("Discount:", discount)
print("Final Bill:", final_bill)
```
