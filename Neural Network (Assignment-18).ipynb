```python
df_new = df.drop('Y', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
MMS = MinMaxScaler()
df_scaled = MMS.fit_transform(df_new)
```

```python
# Create a new DataFrame with normalized values
df_normalized = pd.DataFrame(df_scaled, columns=df_new.columns)
df_normalized['Y'] = df['Y']
```

```python
df_normalized.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Features and labels
X = df_normalized.drop('Y', axis=1).values
y = df_normalized['Y'].values
```

```python
# Encode labels (letters) into numeric categories
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Initialize ANN
model = Sequential()
```

```python
# Input layer + first hidden layer (8 neurons, ReLU activation)
model.add(Dense(8, input_dim=X.shape[1], activation='relu'))
```

```python
# Second hidden layer (optional, 8 neurons, ReLU)
model.add(Dense(8, activation='relu'))
```

```python
# Output layer (number of classes, softmax activation)
num_classes = len(label_encoder.classes_)
model.add(Dense(num_classes, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Model summary
model.summary()
```

```python
# Train the model
model_train = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
# Make predictions
y_pred = model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1)
```

```python
from kerastuner import HyperModel
from kerastuner.tuners import RandomSearch
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Define a HyperModel
class ANNHyperModel(HyperModel):
    def __init__(self, input_dim, num_classes):
        self.input_dim = input_dim
        self.num_classes = num_classes
    
    def build(self, hp):
        model = Sequential()
        
        # Number of hidden layers (1-3)
        for i in range(hp.Int('num_layers', 1, 3)):
            model.add(Dense(
                units=hp.Int(f'units_{i}', min_value=8, max_value=64, step=8),
                activation=hp.Choice('activation', ['relu', 'tanh'])
            ))
        
        # Output layer
        model.add(Dense(self.num_classes, activation='softmax'))
        
        # Compile the model with tunable learning rate
        model.compile(
            optimizer=tf.keras.optimizers.Adam(
                learning_rate=hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])
            ),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        return model
```

```python
# Create HyperModel instance
hypermodel = ANNHyperModel(input_dim=X_train.shape[1], num_classes=num_classes)
```

```python
hypermodel
```

```python
# Define tuner
tuner = RandomSearch(
    hypermodel,
    objective='val_accuracy',
    max_trials=2,  # number of hyperparameter combinations to try
    executions_per_trial=2,  # average result over multiple runs
    directory='hyperparam_tuning',
    project_name='sonar_ANN'
)
```

```python
tuner.search_space_summary()
```

```python
tuner.search(
    X_train, y_train,
    epochs=20,
    validation_split=0.2,
    batch_size=32,
    verbose=1
)
```

```python
# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
# Evaluate on test set
loss, accuracy = best_model.evaluate(X_test, y_test)
print("Best Test Accuracy:", accuracy)
```

```python
# Get best hyperparameters
best_hyperparameters = tuner.get_best_hyperparameters(num_trials=1)[0]
print("Best Hyperparameters:", best_hyperparameters.values)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# For default model
y_pred_default = model.predict(X_test)
y_pred_default_classes = y_pred_default.argmax(axis=1)
```

```python
# For tuned model
y_pred_tuned = best_model.predict(X_test)
y_pred_tuned_classes = y_pred_tuned.argmax(axis=1)
```

```python
# Function to calculate metrics
def evaluate_model(y_true, y_pred_classes, model_name="Model"):
    acc = accuracy_score(y_true, y_pred_classes)
    precision = precision_score(y_true, y_pred_classes, average='weighted')
    recall = recall_score(y_true, y_pred_classes, average='weighted')
    f1 = f1_score(y_true, y_pred_classes, average='weighted')
    
    print(f"--- {model_name} Evaluation ---")
    print("Accuracy:", round(acc, 4))
    print("Precision:", round(precision, 4))
    print("Recall:", round(recall, 4))
    print("F1-score:", round(f1, 4))
    print("\nClassification Report:\n", classification_report(y_true, y_pred_classes))
```

```python
# Evaluate default model
evaluate_model(y_test, y_pred_default_classes, "Default ANN")

# Evaluate tuned model
evaluate_model(y_test, y_pred_tuned_classes, "Tuned ANN")
```
